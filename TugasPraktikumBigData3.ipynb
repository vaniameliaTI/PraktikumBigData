```python
with open('data.txt', 'r') as file:
    texts = file.readlines()

def map_function(text):
    for world in text.split():
        yield (world, 1)
mapped_results = []
for line in texts:
    mapped_results.extend(list(map_function(line)))

print(mapped_results)

from collections import defaultdict

def reduce_function(pairs):
    result = defaultdict(int)
    for word, count in pairs:
        result[word] += count
    return result
reduced_results = reduce_function(mapped_results)
for word, count in reduced_results.items():
    print(f'{word}: {count}')
```

```python

nilai = [85, 77,95, 84, 99]

# Map Function
def map_function(values):
    mapped_nilai = []
    for value in values:
        mapped_nilai.append(("Rata-rata", (value, 1)))
    return mapped_nilai

# Reduce Function
def reduce_function(mapped_nilai):
    total_sum = 0
    total_count = 0
    for key, (value, count) in mapped_nilai:
        total_sum += value
        total_count += count
    average = total_sum / total_count
    return ("Rata-rata", average)

# Proses MapReduce
mapped_nilai = map_function(nilai)
result = reduce_function(mapped_nilai)
print(result)
```
